```python
import pandas as pd

#file_name = input()
file_name = 'd.smokedetector_t/S3_aim1381_sd_devc.csv'
sd=pd.read_csv(file_name, encoding='UTF-8',header=None)
sd.columns = sd.loc[1]
sd_df=sd.copy()

sd_colnames=sd_df.columns[sd_df.columns.str.contains("SD")].tolist()
sd_df=sd_df[sd_colnames].loc[2:]
sd_df=sd_df.reset_index()
sd_df['index'] = range(0,401,1)
sd_df.rename(columns={'index': 'Time(s)'}, inplace=True)
```

```python
sd_df
```

```python
sd_df_melt=sd_df.melt(id_vars=['Time(s)'] ,var_name='smoke detector', value_name='soot mass')
sd_df_melt = sd_df_melt.astype({'soot mass':float})
sd_df_tidy = sd_df_melt[sd_df_melt['soot mass'] > 0].copy()
sd_df_tidy
```

```python
sd_time_sorted=sd_df_tidy.sort_values(by='Time(s)').groupby(by='smoke detector').min()[['Time(s)']].sort_values(by='Time(s)')
sd_time_sorted
```

```python
sd_time_sorted.to_csv(file_name.split("_")[0]+'_tidy data')
```
