```python
import pyAgrum as gum
import pyAgrum.lib.dynamicBN as gdyn
import pyAgrum.lib.notebook as gnb
import IPython

twodbn = gum.BayesNet()
a0, b0, c0, at, bt, ct = [twodbn.add(gum.LabelizedVariable(s,s,2)) for s in ["a0", "b0", "c0", "at", "bt", "ct"]]
d0, dt = [twodbn.add(gum.LabelizedVariable(s,s,2)) for s in ["d0", "dt"]]

twodbn.addArc(a0,b0)
twodbn.addArc(c0,d0)
twodbn.addArc(c0,at)
twodbn.addArc(a0,at)
twodbn.addArc(a0,bt)
twodbn.addArc(a0,dt)
twodbn.addArc(b0,bt)
twodbn.addArc(b0,ct)
twodbn.addArc(c0,ct)
twodbn.addArc(d0,ct)
twodbn.addArc(d0,dt)

twodbn.addArc(at,ct)
twodbn.generateCPTs()

print(twodbn)
```

```python
#for var in twodbn.nodes():
#    print(twodbn.cpt(var))

gdyn.showTimeSlices(twodbn)
```

```python
T=30
dbn = gdyn.unroll2TBN(twodbn, T)
gdyn.showTimeSlices(dbn,size=30)
```

```python
gnb.flow.clear()
for i in range(T):
    gnb.flow.add_html(gnb.getPosterior(dbn, target="a{}".format(i), evs={}), "$P(a{})$".format(i))
gnb.flow.display()
```

```python
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = (10,2)
gdyn.plotFollow(["a","b","c","d"], twodbn, T=51, evs = {'a0':1})
```
